```python
import pandas as pd
from Classes.Database import Database
```

```python
engine = Database()
```

```python
allText = pd.read_sql("SELECT tweet_id, text From tweetCoords", engine.myDB)
```

```python
myVector = dict()
for text in allText.text:
    for word in text.split(' '):
        if word.strip() != '':
            try:
                myVector[word.lower()] += 1
            except:
                myVector[word.lower()] = 1
```

```python
mk, mv = 0,0
for k,v in myVector.items():
    if v > mv:
        mv = v
        mk = k
print mk, mv
```

```python
for key, value in sorted(myVector.iteritems(), key=lambda (k,v): (v,k), reverse=True):
    if(value > 10):
        print "%s: %s" % (key, value)
```
